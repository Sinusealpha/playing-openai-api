```python
import transformers

from transformers import pipeline
# pipeline is the most high level api to work simply with models.
```

### sentiment analysis piepline:

it's a way to understand and classify the words/sentences meaning.

look the following example and change its sentence and re-run it.

**tips:** that 0.9835442900657654 (or similar numbers) you see are the confidence of classifications.

```python
classifier = pipeline("sentiment-analysis")
classifier('this is sina who is happy that there is somehting called hf!')
```

### zero shot (sentiment) analysis

it allows us to classify based on some provided labels and check the relation between the text and those labels.


```python
classifier = pipeline("zero-shot-classification")
classifier("this document is totally came from transformers document and i'm learning a lot by writing it.",
           candidate_labels = ["education", "politics", "business"])
```

### text generation

```python
# we can specify the model we are using for generating!
generator = pipeline("text-generation", model = "distilgpt2")
generator("hi this is sina moradi and i'm gonna",
          max_length = 40,
          num_return_sequences = 3)
# max lenght is the size of generated output.
# num_return_sequences is the numbers of outputs.
```

### fill-mask pipeline

here we want to find the most likely words which have been missed in the sentence.

```python
unmasker = pipeline("fill-mask")

unmasker("hi this is nima moradi and i'm interested in <mask> because it's too fun to hang out with her.",
         top_k = 3)
# top_k is the numbers of outputs you want for the masked word.
```

### ner pipeline

now we want to find the entities (names, locations, etc.) in a given sentence.

```python
ner = pipeline("ner", grouped_entities = True)
ner("my name is Sina Moradi, i love Math and i want to work on NLP's application in Healthcare industry.")
```

### question answering pipeline


```python
question_answerer = pipeline("question-answering")
question_answerer(
    question = "is math my love or biology?",
    context = "i love math lessons more than biology and i love physics more than chemistry."
)
```

### translation pipeline


```python
x = str(input("please enter your sentence: "))

translator = pipeline("translation", model = "Helsinki-NLP/opus-mt-fr-en")
translator(x)
```
